Los datos de tiempos de produccion en una tienda de caramelos cada empleado se comporta de manera exponencial con una media de 10 minutos/tandada 

```python
from math import log
import matplotlib.pyplot as plt
import pandas as pd
import random
import math
import numpy as np
```

```python
def transformada_inversa(numeros, a):
    transf_inv=[]
    n_transf_inv=[]
    densidad=[]
    f=[]
    numeros=sorted(numeros)
    for u in numeros:
        aux=-(log(1-u)/a)
        transf_inv.append(aux)

        aux=a*math.exp(-a*u)
        densidad.append(aux)



    df = pd.DataFrame(list(zip(numeros,transf_inv)), columns = ['ri','Tiempo de produccion'])
    print(df)
    plt.figure(figsize=(16,6))
    plt.subplot(1, 2, 1)
    plt.plot(numeros, densidad)
    plt.title("Funcion de la densidad")
    plt.subplot(1, 2, 2)
    plt.plot(numeros, transf_inv)
    plt.title("Funcion de la probabilidad acumulada")
```
